# Imersão

### Importando bibliotecas

```python
import pandas as pd
import plotly.express as px
```

Importando arquivos

```python
df_principal = pd.read_excel("/content/Tabela de ações pura.xlsx", sheet_name="Principal")
df_principal
```

```python
df_totalAcoes = pd.read_excel("/content/Tabela de ações pura.xlsx", sheet_name="Total_de_acoes")
df_totalAcoes
```

```python
df_ticker = pd.read_excel("/content/Tabela de ações pura.xlsx", sheet_name="Ticker")
df_ticker
```

```python
df_chatgpt = pd.read_excel("/content/Tabela de ações pura.xlsx", sheet_name="chatGPT")
df_chatgpt
```

Atualizando o data frame apenas com as colunas que iremos utilizar

```python
df_principal = df_principal[['Ativo', 'Data', 'Último (R$)', 'Var. Dia (%)']].copy()
df_principal
```

 Renomeando colunas último (R$) e Var. Dia (%)

```python
df_principal = df_principal.rename(columns = {'Último (R$)': 'valorFinal','Var. Dia (%)': 'variacaoDiaPCT'}).copy()
df_principal
```

Criando coluna de variação do dia em porcentagem (variacao_pct)

```python
df_principal['variacao_pct'] = df_principal['variacaoDiaPCT']/100
df_principal.head(5)
```

Criando a coluna 'VALOR INICIAL'

```python
df_principal['valorInicial'] = df_principal['valorFinal'] / (df_principal['variacao_pct'] + 1)
df_principal.head(5)
```

Criando coluna Quantidade de ações (qntd_acoes)

```python
df_principal = df_principal.merge(df_totalAcoes, left_on= 'Ativo', right_on = 'Código', how='left')
df_principal.head(5)
```

Removendo coluna código (duplicidade de valores)

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal.head(5)
```

Criando coluna (Variação dia ($))

```python
df_principal['VariacaoDiaEmReais'] = (df_principal['valorFinal'] - df_principal['valorInicial']) * df_principal['Qtde. Teórica']
df_principal.head(5)
```

Transformando floats (números decimais) para duas casas decimais

```python
pd.options.display.float_format = '{:.2f}'.format
df_principal.head(5)
```

Criando coluna Resultado (Subiu, Desceu, Estável)

```python
df_principal['Resultado'] = pd.Series(['SUBIU' if x > 0 else 'ESTÁVEL' if x == 0 else 'DESCEU' for x in df_principal['VariacaoDiaEmReais']])
df_principal.head(5)
```

Criando coluna (Nome da empresa)

```python
df_principal = df_principal.merge(df_ticker, left_on= 'Ativo', right_on = 'Ticker', how='left')
df_principal.head(5)
```

Removendo coluna Ticker (duplicidade de valor)

```python
df_principal = df_principal.drop(columns=['Ticker'])
df_principal.head(5)
```

Adicionando colunas Idade, cat_Idade, Segmento



```python
df_principal = df_principal.merge(df_chatgpt, left_on= 'Nome', right_on = 'Nome da Empresa', how='left')
df_principal.head(5)
```

```python
df_principal = df_principal.drop(columns = ['Nome da Empresa'])
df_principal.head(5)
```

```python
df_principal = df_principal.rename(columns = {'Idade (em anos)': 'Idade'}).copy()
df_principal.head(5)
```

```python
df_principal['categoria_idade'] = pd.Series(['MAIS DE CEM ANOS' if x > 100 else 'MENOS DE 50 ANOS' if x < 50 else 'ENTRE 50 E 100 ANOS' for x in df_principal['Idade']])
df_principal.head(10)
```

(Qtde. Teórica) de float para int (número inteiro)

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns = {'Qtde. Teórica': 'qntdAcoes'}).copy()
df_principal
```

```python
df_principal = df_principal.drop_duplicates()
df_principal
```

#Análises

```python
maior = df_principal['VariacaoDiaEmReais'].max()
idmaior = df_principal['VariacaoDiaEmReais'].idxmax()
empresaMaior = df_principal.loc[idmaior, 'Nome']
print(f"A empresa que mais subiu foi a {empresaMaior} com R${maior}")
```

```python
menor = df_principal['VariacaoDiaEmReais'].min()
idmenor = df_principal['VariacaoDiaEmReais'].idxmin()
empresamenor = df_principal.loc[idmenor, 'Nome']
print(f"A empresa que mais desceu foi a {empresamenor} com  R${menor}")
```

```python
qntdTotal = df_principal['Resultado'].count()
mediaTotal = df_principal['VariacaoDiaEmReais'].mean()
print(f'Média de todas as empresas foi de: R${mediaTotal}')
print(qntdTotal)
```

```python
qntdSubiu = df_principal[df_principal['Resultado'] == 'SUBIU']['VariacaoDiaEmReais'].count()
mediaSubiu = df_principal[df_principal['Resultado'] == 'SUBIU']['VariacaoDiaEmReais'].mean()
print(f'Quantidade das empresas que subiram é de {qntdSubiu}. E média dessas empresas: {mediaSubiu}')
```

```python
qntdDesceu = df_principal[df_principal['Resultado'] == 'DESCEU']['VariacaoDiaEmReais'].count()
mediaDesceu = df_principal[df_principal['Resultado'] == 'DESCEU']['VariacaoDiaEmReais'].mean()
print(f'Quantidade das empresas que subiram é de {qntdDesceu}. E média dessas empresas: {mediaDesceu}')
```

Criando data frame para mostrar apenas dados das empresas que subiram

```python
df_principal_subiu = df_principal[df_principal['Resultado'] == 'SUBIU']
df_principal_subiu
```

Criando data frame para mostrar apenas dados por segmento

```python
df_segmento = df_principal_subiu.groupby('Segmento')['VariacaoDiaEmReais'].sum().reset_index()
df_segmento
```

Criando data frame para fazer uma análise por saldo

```python
df_saldo = df_principal.groupby('Resultado')['VariacaoDiaEmReais'].sum().reset_index()
df_saldo
```

```python
fig_varDia = px.bar(df_saldo, x='Resultado', y='VariacaoDiaEmReais', text='VariacaoDiaEmReais', title='Variação diária R$')
fig_varDia.update_layout(width = 700)
fig_varDia.show()
```
